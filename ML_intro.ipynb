```python
from scipy import stats

import matplotlib.pyplot as plt
```

```python
n = 1000
p = 0.5
```

```python
x = list(range(0, n+1))
x
```

```python
y = stats.binom.pmf(x, n, p)
y
```

find the probability of getting 4 heads 

```python
stats.binom.pmf(4, n, p)
```

```python
stats.binom.pmf(2, n, p)
```

```python
z = stats.poisson.pmf(x, n*p)
z
```

```python
plt.plot(x,z, label="bino")
```

```python
# plt.plot(x, y, label="binomial")
plt.plot(x,z, label="bino")
plt.plot(x, z, label="poisson")
plt.ylabel("probability")
plt.xlabel("random variable")
plt.legend()
```

```python
a = stats.norm.pdf(x, n*p, n*p*(1-p))
a
```

```python
plt.plot(x, a)
```
